## Create a Pandas DF by reading the 'block_9.csv' file 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("block_9_data.csv")
df.head()
```

## Create a new DF which reports the sum of reviews by year

```python
a = df.copy()
a.loc[:,'count'] = 1
a = a[['review_year','count']]
a = pd.DataFrame(a.groupby('review_year')['count'].aggregate(np.sum)).rename(columns={'count':'sum_reviews'})
a
```

## Create a new DF which reports the sum of reviews by month and year

```python
b = df.copy()
b.loc[:,'count'] = 1
b = b[['review_year','review_month','count']]
b = pd.DataFrame(b.groupby(['review_year','review_month'])['count'].aggregate(np.sum)).rename(columns={'count':'sum_reviews'})
b
```

## Create a new DF which reports the mean and standard deviation in the number of  reviews by month

```python
c = b.copy().reset_index()
left = pd.DataFrame(c.groupby('review_month')['sum_reviews'].aggregate(np.mean)).rename(columns={'sum_reviews':'mean'}).reset_index()
right = pd.DataFrame(c.groupby('review_month')['sum_reviews'].aggregate(np.std)).rename(columns={'sum_reviews':'std'}).reset_index()
c = left.merge(right, how='left', on='review_month')
c
```
